```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
df.pivot_table(values='D',
               index=['A','B'],
               columns=['C'],
               aggfunc=np.sum
              )
```

```python
pd.pivot_table(df,
         values='D',
         index=['A', 'B'],
         columns='C',
         aggfunc=np.sum,
         fill_value=0
        )
```

```python
pd.pivot_table(df,
               values=['D', 'E'],
               index=['A','C'],
               aggfunc={
                   'D': np.mean,
                   'E': np.mean
               }
              )
```

```python
pd.pivot_table(df,
               values=['D','E'],
               index=['A','C'],
               aggfunc={'D':np.mean,'E':[min, max, np.mean]}
              )
```
